```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configuracion general de pyplot 
# (Revisar Doc)
plt.rcParams["figure.figsize"]= (4,3)
plt.rcParams["figure.dpi"] = 150
```

```python
valoresx = np.arange(0,10,0.1)
```

```python
valoresx.shape
```

```python
# plt.plot(valoresx)
plt.plot(valoresx, np.sin(valoresx))
```

```python
plt.plot(valoresx, np.sin(valoresx))
plt.plot(valoresx, np.cos(valoresx))
plt.legend(["Seno", "Coseno"])
plt.xlabel("Valores en X")
plt.ylabel("Valores en Y")
```
